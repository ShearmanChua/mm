```python
import requests
import os
import base64
import json
```

```python
os.listdir('../data/trump')
```

```python
b64_list = []
for i in os.listdir('../data/trump'):
    full_path = os.path.join('../data/trump', i)
    print(full_path)
    with open(full_path, "rb") as f:
        im_bytes = f.read()
    im_b64 = base64.b64encode(im_bytes).decode("utf8")
    b64_list.append({'image':im_b64})
    
headers = {'Content-type': 'application/json',
           'Accept': 'text/plain'}
payload = json.dumps({"images": b64_list})

r_fn = requests.post(
    f"http://faceid:8000/embedding", data=payload, headers=headers)
res_fn = json.loads(r_fn.text)
avg3 = res_fn['emb']
```
